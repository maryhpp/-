```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/robot.jpg')
height, width = img.shape[:2]
def show_imgs(img,res):
    plt.figure(figsize=(20,6))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
    plt.show()
```

# 1-改变图像大小(resize)

```python
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
show_imgs(img,res)
```

# 2-水平翻转

```python
res =cv2.flip(img,1)
show_imgs(img,res)
```

# 3-垂直翻转

```python
res =cv2.flip(img,-1)
show_imgs(img,res)
```

# 4-平移

```python
blank=np.zeros([height,width,3])
blank[100:,100:]=img[100:,100:]
trans_img=blank.astype('uint8')
show_imgs(img,trans_img)
```
