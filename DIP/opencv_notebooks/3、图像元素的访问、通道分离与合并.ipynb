## 1. 图像元素的访问
### 1.1 导入相关库文件

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (15.0, 8.0)
```

### 1.2. 展示图片

```python
img = cv2.imread("./images/dog.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# 去除x,y轴刻度
#plt.xticks([])
#plt.yticks([])
plt.show()
```

## 2.分离、合并通道
可以使用OpenCV自带的split函数，也可以直接操作numpy数组来分离通道。   
### 2.1 分离通道

```python
b, g, r = cv2.split(img)
```

其中split返回RGB三个通道，如果只想返回其中一个通道，可以这样：

```python
b = cv2.split(img)[0]
g = cv2.split(img)[1]
r = cv2.split(img)[2]
```

### 2.2 通道合并
同样，通道合并也有两种方法。第一种是OpenCV自带的merge函数，如下：

```python
merged = cv2.merge([b,g,r])
```

接着是NumPy的方法：

```python
mergedByNp = np.dstack([b,g,r]) 
```

# 3.获取子图

```python
y1,y2,x1,x2=0,60,260,350
img_crop=img[y1:y2,x1:x2]
plt.imshow(cv2.cvtColor(img_crop, cv2.COLOR_BGR2RGB))
```
