## 2. 读取并显示图像

```python
img = cv2.imread("./images/robot.jpg")
# 将BGR图像转变为RGB图像以打印
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
# 去除x,y轴刻度
plt.xticks([])
plt.yticks([])
plt.show()
```

## 4. 保存图像
使用cv2.imwrite来保存图像

```python
cv2.imwrite("./images/robot2.jpg", img)
```

第一个参数是保存的路径及文件名，第二个是图像矩阵。其中，imwrite()有个可选的第三个参数，如下：

```python
cv2.imwrite("./images/robot2.jpg", img, [int(cv2.IMWRITE_JPEG_QUALITY), 5])
```

第三个参数针对特定的格式： 对于JPEG，其表示的是图像的质量，用0-100的整数表示，默认为95。 注意，cv2.IMWRITE_JPEG_QUALITY类型为Long，必须转换成int。
对于PNG，第三个参数表示的是压缩级别。cv2.IMWRITE_PNG_COMPRESSION，从0到9,压缩级别越高，图像尺寸越小。默认级别为3：

```python
cv2.imwrite("./images/robot.png", img, [int(cv2.IMWRITE_PNG_COMPRESSION), 6]) 
```
